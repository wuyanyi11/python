```python
#第二题
import pandas as pd
from pandas import Series, DataFrame
data = {'语文':[68,95,98,90,80],'数学':[65,76,86,88,90],'英语':[30,98,88,77,90]}
df = DataFrame(data,index=['张飞','关羽','刘备','典韦','许褚'])
print(df)
print('语文、英语、数学中的平均成绩')
print(df.mean())
print('语文、英语、数学中的最小成绩')
print(df.min())
print('语文、英语、数学中的最大成绩')
print(df.max())
print('语文、英语、数学中的方差')
print(df.var())
print('语文、英语、数学中的标准差')
print(df.std())
#data={data,'总分'='语文'+'数学'+'英语'}
#df = DataFrame(data,index=['张飞','关羽','刘备','典韦','许褚'])
df["总分"] = df["语文"] + df["数学"] + df["英语"]
print(df.sort_values('总分',ascending=False))

```
